```python
df=df.drop_duplicates()     # drop duplicate records
df.shape
```

```python
df["Supplier"]=df["Supplier"].fillna(df["Supplier"].mode()[0])    # fill missing values with mode of the rows
df["Item Type"]=df["Item Type"].fillna(df["Item Type"].mode()[0])
df["Retail Sales"]=df["Retail Sales"].fillna(df["Retail Sales"].mode()[0])
```

```python
df.isnull().sum()   #treated missing values
```

```python
def date(row):                                    
    y=row["Year"]
    z=row["Month"]
    date=f"{y} - {z}"
    return date


df["Date"]=df.apply(date,axis=1)

df["Total Sales"]=df["Retail Sales"]+df["Warehouse Sales"]
df=df.drop(["Year","Month"],axis=1)
df
```

```python
df["Supplier"]=df["Supplier"].str.strip().str.title()        # 1st letter capital in each word in the column 
df["Item Description"]=df["Item Description"].str.strip().str.title()  # 1st letter capital in each word in the column 
df["Item Type"]=df["Item Type"].str.strip().str.title()    # 1st letter capital in each word in the column 
df
```

```python
df[df["Total Sales"]<=0]   # sales cant be negative checking <=0 values
```

```python
df = df[(df["Total Sales"] > 0) & (df["Warehouse Sales"] > 0) & (df["Retail Sales"] > 0) & (df["Retail Transfers"] > 0)]
df
```

```python
df.describe()
```

```python
 #converted date column to date time format # errors="coerce" means avoided missing values and other format data

df["Date"]=pd.to_datetime(df["Date"],errors="coerce")  
df
```

```python
#check outliers

import seaborn as sns
plt.figure(figsize=(10,4))
sns.boxplot(data=df[['Retail Sales', 'Warehouse Sales',"Total Sales"]])
plt.title("Boxplot of Sales Columns")
plt.show()
```

```python
Q1=df["Total Sales"].quantile(0.25)
Q3=df["Total Sales"].quantile(0.75)

IQR=Q3-Q1
ll=Q1-(1.5*IQR)
ul=Q3+(1.5*IQR)

outliers=df[(df["Total Sales"]<ll) | (df["Total Sales"]>ul)]   
len(outliers) 
```

```python
df=df[(df["Total Sales"]>ll) & (df["Total Sales"]<ul)]   # remove outliers 
df
```

```python
# Check if the data is normal
df["Total Sales"].skew()     # right skewed ----   (normal range: -1<skewness<1 )
```

```python
# Visual Check
sns.histplot(df["Total Sales"],kde=True) 
plt.show()
```

```python
# Feature Transformation -- Right-skewed data normalized using log transformation

df["log_total_sales"] = np.log1p(df["Total Sales"])
df["log_total_sales"]
```
